```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate x values from 1 to just below 100
x = np.linspace(1, 99.9, 1000)

# Compute y using the formula
y = (x * 100) / (100 - x)
```

```python
# Create the plot with higher DPI for HD quality
plt.figure(figsize=(8, 8), dpi=300)
plt.plot(x, y, color='royalblue', linewidth=2)

# Adding title and labels
plt.title('Stock Recovery Gain Curve', fontsize=16, pad=20)
plt.xlabel('Percentage Drop in Stock (%)', fontsize=14, labelpad=15)
plt.ylabel('Gain Required to Break Even (%)', fontsize=14, labelpad=15)

# Adding grid for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Set axis limits for symmetry
plt.xlim(0, 100)
plt.ylim(0, 600)

# Make the aspect ratio of x and y axis equal to emphasize asymmetry
plt.gca().set_aspect('equal', adjustable='box')

# Adjust layout to ensure there's enough space for labels and title
plt.tight_layout()

# Show the plot
plt.show()
```
